# Data 불러오기

```python
import pandas as pd
import numpy as np
```

```python
xls = pd.ExcelFile("C:\\Users\\renz\\Desktop\\Concrete_Data.xls")
org_data = xls.parse('Sheet1')
```

# Data 전처리

```python
org_data = org_data.rename(columns={'Cement (component 1)(kg in a m^3 mixture)':"cement",
       'Blast Furnace Slag (component 2)(kg in a m^3 mixture)':"furnace_slag",
       'Fly Ash (component 3)(kg in a m^3 mixture)':"fly_ash",
       'Water  (component 4)(kg in a m^3 mixture)':"water",
       'Superplasticizer (component 5)(kg in a m^3 mixture)':"super_plasticizer",
       'Coarse Aggregate  (component 6)(kg in a m^3 mixture)':"coarse_agg",
       'Fine Aggregate (component 7)(kg in a m^3 mixture)':"fine_agg", 'Age (day)':"age",
       'Concrete compressive strength(MPa, megapascals) ':"compressive_strength"})
```

```python
data1=org_data.copy()
data1.head()
```

```python
cnt = 0;
for i in data1['age']:
    if i < 28 :
            data1['age'][cnt] = 'class_1';
    elif i == 28 :
            data1['age'][cnt] = 'class_2';
    else :
            data1['age'][cnt] = 'class_3';
    cnt+=1;
data1
```

```python
data1.head()
```

```python
dummies = pd.get_dummies(data1['age'], prefix='age')
dummies.head()
```

```python
com = org_data.drop(['cement', 'furnace_slag', 'fly_ash', 'water', 'super_plasticizer',
       'coarse_agg', 'fine_agg','age'], axis=1)
com.head()
```

```python
data1=data1.drop(['age','compressive_strength'], axis=1)
data1.head()
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler().fit(data1)
data1_scaled = pd.DataFrame(scaler.transform(data1),
                               columns = data1.columns)
data1_scaled.head()
```

```python
data1=data1_scaled.join(dummies)
data1.head()
```

```python
data1=data1.join(com)
data1.head()
```

# Train, Test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data1[data1.columns[:-1]],
                                                    data1[[data1.columns[-1]]],
                                                    test_size = 0.2,
                                                   random_state=1)
```

```python
y_train.head()
```

```python
X_train_scaled = X_train
```

```python
X_train_scaled.head()
```

# Cross Validation with default parameter

```python
import warnings
warnings.filterwarnings("ignore")

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
%matplotlib inline
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import KFold, cross_val_score

Model = []
RMSE = []
R_sq = []
cv = KFold(5, random_state = 1)

def input_scores(name, model, X, y):
    Model.append(name)
    RMSE.append(np.sqrt((-1) * cross_val_score(model, X, y, cv=cv, 
                                               scoring='neg_mean_squared_error')))
    R_sq.append(cross_val_score(model, X, y, cv=cv, scoring='r2'))
```

```python
y_train_raveled = np.ravel(y_train)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.cross_decomposition import PLSRegression as PLS
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

names = ['Linear Regression', 'Ridge Regression', 'Lasso Regression',
         'ElasticNet', 'K Neighbors Regressor', 'svr', 'PLS']
models = [LinearRegression(), Ridge(), Lasso(),
          ElasticNet(), KNeighborsRegressor(), SVR(), PLS()]

for name, model in zip(names, models):
    input_scores(name, model, X_train_scaled, y_train_raveled)
```

```python
evaluation = pd.DataFrame({'Model': Model,
                           'RMSE': RMSE,
                           'R-Squared': R_sq})
print("FOLLOWING ARE THE TRAINING SCORES: ")
evaluation
```

```python
RMSE_default = [evaluation.loc[i,'RMSE'] for i in range(0,7)]
Rsq_default = [evaluation.loc[i,'R-Squared'] for i in range(0,7)]
```

```python
from pandas import DataFrame, Series
```

```python
RMSE_default_df = DataFrame(RMSE_default, columns=['Fold_'+str(i) for i in range(1,6)])
Rsq_default_df = DataFrame(Rsq_default, columns=['Fold_'+str(i) for i in range(1,6)])
```

```python
model_index = evaluation['Model'].ravel()
model_index
```

```python
mpl.rcdefaults()
for i in range(1,6):
    plt.plot(RMSE_default_df['Fold_'+str(i)], model_index)
plt.xlabel('RMSE')
plt.grid(linestyle=':')
plt.title('RMSE using Default Parameter')
plt.show()
```

```python
mpl.rcdefaults()
for i in range(1,6):
    plt.plot(Rsq_default_df['Fold_'+str(i)], model_index)
plt.xlabel('R-Squared')
plt.grid(linestyle=':')
plt.title('R-Squared using Default Parameter')
plt.show()
```

# Find Best Parameter

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'alpha':[0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0],
             'max_iter':range(100, 1000, 10)}
fbp_rdg = GridSearchCV(Ridge(random_state=1), 
                   param_grid = param_grid, scoring='r2', 
                   cv=cv).fit(X_train_scaled, y_train_raveled)
print(fbp_rdg.best_params_)
print("Ridge R-Squared:",fbp_rdg.best_score_)
```

```python
param_grid = {'alpha':[0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0],
             'max_iter':range(100, 1000, 10)}
fbp_lss = GridSearchCV(Lasso(random_state=1), 
                   param_grid = param_grid, scoring='r2', 
                   cv=cv).fit(X_train_scaled, y_train_raveled)
print(fbp_lss.best_params_)
print("Lasso R-Squared:",fbp_lss.best_score_)
```

```python
param_grid = {'alpha':[0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0],
             'max_iter':range(100, 1000, 10)}
fbp_ela = GridSearchCV(ElasticNet(random_state=1), 
                   param_grid = param_grid, scoring='r2', 
                   cv=cv).fit(X_train_scaled, y_train_raveled)
print(fbp_ela.best_params_)
print("ElasticNet R-Squared:",fbp_ela.best_score_)
```

```python
param_grid = {'leaf_size':[10, 100, 10],
             'n_neighbors':range(1, 11, 1), 'p':[1,2]}
fbp_knr = GridSearchCV(KNeighborsRegressor(), 
                   param_grid = param_grid, scoring='r2', 
                   cv=cv).fit(X_train_scaled, y_train_raveled)
print(fbp_knr.best_params_)
print("KNeighborsRegressor R-Squared:",fbp_knr.best_score_)
```

```python
param_grid = {'C':[0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0],
              'epsilon':[0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0],
             'kernel':['linear','poly','rbf','sigmoid']}
fbp_svr = GridSearchCV(SVR(), 
                   param_grid = param_grid, scoring='r2', 
                   cv=cv).fit(X_train_scaled, y_train_raveled)
print(fbp_svr.best_params_)
print("SVR R-Squared:",fbp_svr.best_score_)
```

```python
param_grid = {'max_iter':range(100, 1000, 10),
              'n_components':range(1, 6, 1)}
fbp_pls = GridSearchCV(PLS(), 
                   param_grid = param_grid, scoring='r2', 
                   cv=cv).fit(X_train_scaled, y_train_raveled)
print(fbp_pls.best_params_)
print("SVR R-Squared:",fbp_pls.best_score_)
```

# Cross Validation with best parameter

```python
Model = []
RMSE = []
R_sq = []
cv = KFold(5, random_state = 1)

def input_scores_bestparam(name, model, X, y):
    Model.append(name)
    RMSE.append(np.sqrt((-1) * cross_val_score(model, X, y, cv=cv, 
                                               scoring='neg_mean_squared_error')))
    R_sq.append(cross_val_score(model, X, y, cv=cv, scoring='r2'))
```

```python
names = ['Linear Regression', 'Ridge Regression', 'Lasso Regression',
         'ElasticNet', 'K Neighbors Regressor', 'svr', 'PLS']
models = [LinearRegression(), 
          fbp_rdg.best_estimator_, 
          fbp_lss.best_estimator_,
          fbp_ela.best_estimator_, 
          fbp_knr.best_estimator_,
          fbp_svr.best_estimator_,
          fbp_pls.best_estimator_]

for name, model in zip(names, models):
    input_scores_bestparam(name, model, X_train_scaled, y_train_raveled)
```

```python
evaluation_last = DataFrame({'Model': Model,
                           'RMSE': RMSE,
                           'R-Squared': R_sq})
print("FOLLOWING ARE THE TRAINING SCORES: ")
evaluation_last
```

```python
RMSE_best=[evaluation_last.loc[i,'RMSE'] for i in range(0,7)]
Rsq_best=[evaluation_last.loc[i,'R-Squared'] for i in range(0,7)]
```

```python
RMSE_best_df = DataFrame(RMSE_best, columns=['Fold_'+str(i) for i in range(1,6)])
Rsq_best_df = DataFrame(Rsq_best, columns=['Fold_'+str(i) for i in range(1,6)])
```

```python
index_best = evaluation_last['Model'].ravel()
```

```python
mpl.rcdefaults()
for i in range(1,6):
    plt.plot(RMSE_best_df['Fold_'+str(i)], index_best)
plt.xlabel('RMSE')
plt.grid(linestyle=':')
plt.title('RMSE using Best Parameter')
plt.show()
```

```python
for i in range(1,6):
    plt.plot(Rsq_best_df['Fold_'+str(i)], index_best)
plt.xlabel('R-Squared')
plt.grid(linestyle=':')
plt.title('R-Squared using Best Parameter')
plt.show()
```
